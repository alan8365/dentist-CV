```python
from edge.processor import ImageProcessor
import os
import json
from pathlib import Path

import torch

import matplotlib
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
from dotenv import load_dotenv
from scipy.ndimage import rotate
from yolov5.models.common import Detections

load_dotenv()
matplotlib.use('module://matplotlib_inline.backend_inline')

%matplotlib inline
```

```python
data_dir = Path(os.getenv('DATASET_DIR')) / 'phase-2'
data_dir
```

```python
plt.gray()

image_paths = list(data_dir.glob('0000*.jpg'))
image_paths = sorted(image_paths)
image_paths[:30]
```

```python
import json

image_name = image_paths[0]

img = cv.imread(image_name)
label_name = image_name.with_suffix('.json')

with open(label_name, 'r') as f:
    label_file = json.load(f)

shapes = label_file['shapes']
gaps_truth = [np.hstack(shape['points']) for shape in shapes if shape['label'] == 'gap']

plt.imshow(1 - img)
for gap in gaps_truth:
    plt.scatter(*gap)
plt.show()
```
